```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot

df = pd.read_csv('middleSchoolData.csv')
df = df.dropna(how='any', subset=['student_achievement', 'reading_scores_exceed', 'math_scores_exceed',
                                 'avg_class_size', 'acceptances'])

df = (df-df.mean())/df.std()
achievements = (df['student_achievement'] + df['reading_scores_exceed'] + df['math_scores_exceed'])/3


classAchievementCorr = df['avg_class_size'].corr(achievements)
classAcceptanceCorr = df['avg_class_size'].corr(df['acceptances'])

spendingAchievementCorr = df['per_pupil_spending'].corr(achievements)
spendingAcceptanceCorr = df['per_pupil_spending'].corr(df['acceptances'])



print('classAchievementCorr: ' + str(classAchievementCorr))
print('classAcceptanceCorr: ' + str(classAcceptanceCorr))
print('spendingAchievementCorr: ' + str(spendingAchievementCorr))
print('spendingAcceptanceCorr: ' + str(spendingAcceptanceCorr))

plot.scatter(x=df['avg_class_size'], y=achievements)
plot.xlabel('Avg Class Size')
plot.ylabel('Achievements')
plot.title('Avg Class Size vs Achievements')
plot.show()

plot.scatter(x=df['avg_class_size'], y=df['acceptances'])
plot.xlabel('Avg Class Size')
plot.ylabel('Acceptances')
plot.title('Avg Class Size vs Acceptances')
plot.show()

plot.scatter(x=df['per_pupil_spending'], y=achievements)
plot.xlabel('Per Pupil Spending')
plot.ylabel('Acceptances')
plot.title('Per Pupil Spending vs Acceptances')
plot.show()

plot.scatter(x=df['per_pupil_spending'], y=achievements)
plot.xlabel('Per Pupil Spending')
plot.ylabel('Acceptances')
plot.title('Per Pupil Spending vs Acceptances')
plot.show()
```
