```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot

#neither of these have NaN
df = pd.read_csv('middleSchoolData.csv')
df = df.dropna(how='any', subset=['applications', 'acceptances'])

apps = df['applications']
accepts = df['acceptances']
```

```python
correlation = apps.corr(accepts)
```

```python
print(correlation)
```
